```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import Data

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Chapter9_Unsupervised/newhousing.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

## Select X and y

```python
X = dataset.drop('price', axis = 1)
y = dataset[['price']]
```

## Split data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=49)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
lin_reg_score = lin_reg.score(X_test, y_test)
lin_reg_score
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rand_forest = RandomForestRegressor(n_estimators=500, random_state=49)
rand_forest.fit(X_train, y_train)
rand_forest_score = rand_forest.score(X_test, y_test)
rand_forest_score
```

## Scale data for PCA

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_scaled = sc_x.fit_transform(X)
sc_y = StandardScaler()
y_scaled = sc_y.fit_transform(y)
```

## Prep PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
x_pca = pca.fit_transform(X_scaled)
```

## Resplit data

```python
X_train, X_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.2, random_state=49)
```

## Linear Regression with PCA

```python
lin_reg_pca = LinearRegression()
lin_reg_pca.fit(X_train, y_train)
lin_reg_pca_score = lin_reg_pca.score(X_test, y_test)
lin_reg_pca_score
```

## Random Forest with PCA

```python
rand_forest_pca = RandomForestRegressor(random_state=49, n_estimators=500)
rand_forest_pca.fit(X_train, y_train)
rand_forest_pca_score = rand_forest_pca.score(X_test, y_test)
```

## Results

```python
from prettytable import PrettyTable
table = PrettyTable(["Techniques", "Score"])
table.add_rows([
    ["Linear Regression", lin_reg_score],
    ["Random Forest", rand_forest_score],
    ["Linear Regression with PCA", lin_reg_pca_score],
    ["Random Forest with PCA", rand_forest_pca_score]
])

print(table)
```

Best score: Random Forest with PCA
